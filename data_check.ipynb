```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("sampled_data.csv")
```

```python
columns_to_keep = ['category', 'style', 'tops_fit', 'bottoms_fit', 'sleeve_type', 'pattern', 'more_attributes', 'image_url_1', 'brand']
df = df[columns_to_keep]
#drop duplicate rows
df = df.drop_duplicates()
```

```python
for category in df['category'].unique():
    subset = df[df['category'] == category]  # Filter by category
    print(len(subset))
    missing_counts = subset.isna().sum()  # Count NaN values per column

    plt.figure(figsize=(10, 5))
    sns.barplot(x=missing_counts.index, y=missing_counts.values, palette="viridis")
    plt.xticks(rotation=45)
    plt.ylabel("Number of NaN Values")
    plt.xlabel("Columns")
    plt.title(f"Missing Values in {category}")
    plt.show()
```

```python
# Get unique values for sleeve types and fits
unique_sleeve_types = df['sleeve_type'].dropna().unique()
unique_tops_fits = df['tops_fit'].dropna().unique()
unique_bottoms_fits = df['bottoms_fit'].dropna().unique()
unique_attributes = df['more_attributes'].dropna().unique()

# Print results
print("Unique Sleeve Types:")
print(unique_sleeve_types)

print("\nUnique Tops Fit:")
print(unique_tops_fits)

print("\nUnique Bottoms Fit:")
print(unique_bottoms_fits)

print("\nUnique more attributes:")
print(unique_attributes)
```
